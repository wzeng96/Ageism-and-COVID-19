```python
import pandas as pd
import glob

path = r'D:\Download\nov-dec' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame = frame.drop(columns=['web-scraper-order', 'web-scraper-start-url','link', 'link-href'])
frame
```

```python
frame.to_csv('D:/Download/nov-dec/nov-dec_All.csv', index=None, header=True, encoding='utf_8_sig')
```
